```python
import pandas as pd

# Caminho do arquivo CSV
caminho_csv = r'D:\03 - Portfolios\portfolio_diego\analise_aneel_satisfacao_consumidor\dados\indice-aneel-satisfacao-consumidor.csv'

# Leitura do CSV com delimitador ;
df = pd.read_csv(caminho_csv, sep=';', encoding='latin1')

# Renomear colunas para remover espaços e padronizar
df.columns = df.columns.str.strip().str.replace(' ', '_').str.replace('ç', 'c').str.replace('ã', 'a').str.replace('é', 'e')

# Verificando os nomes das colunas ajustados
print("Colunas do DataFrame:\n", df.columns.tolist())

# Verificar as primeiras linhas
print("\nAmostra dos dados:")
print(df.head())

# Verificar informações gerais
print("\nInformações gerais:")
print(df.info())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Caminho para salvar os gráficos
caminho_imagens = r'D:\03 - Portfolios\portfolio_diego\analise_aneel_satisfacao_consumidor\imagens'

plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Ano', color='steelblue')
plt.title('Distribuição de Registros por Ano')
plt.xlabel('Ano')
plt.ylabel('Número de Registros')
plt.tight_layout()
plt.savefig(f'{caminho_imagens}\\distribuicao_por_ano.png')
plt.show()
```

```python
plt.figure(figsize=(10, 10))
participacoes = df['Agente'].value_counts().head(15).sort_values()
sns.barplot(x=participacoes.values, y=participacoes.index, color='orange')
plt.title('Top 15 Agentes com Mais Registros')
plt.xlabel('Número de Participações')
plt.ylabel('Agente')
plt.tight_layout()
plt.savefig(f'{caminho_imagens}\\top_15_agentes_participacoes.png')
plt.show()
```
