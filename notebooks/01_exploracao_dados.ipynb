```python
import matplotlib.pyplot as plt
import seaborn as sns
import os

def salvar_grafico(fig, nome_arquivo, pasta_destino):
    """
    Salva o gráfico (fig) como PNG no caminho especificado.
    """
    caminho_completo = os.path.join(pasta_destino, f"{nome_arquivo}.png")
    fig.savefig(caminho_completo, dpi=300, bbox_inches='tight')
    plt.close(fig)
    print(f"Gráfico salvo em: {caminho_completo}")
```

```python
import pandas as pd

# 1️⃣ Ler o CSV com separador correto
df = pd.read_csv(r'D:\03 - Portfolios\portfolio_diego\analise_aneel_satisfacao_consumidor\dados\indice-aneel-satisfacao-consumidor.csv',
                 sep=';', encoding='latin1')
df.columns = df.columns.str.strip()

# 2️⃣ Gerar o gráfico
fig, ax = plt.subplots(figsize=(12, 6))
sns.histplot(df['MdaIndicadorSatisfacao'], bins=20, kde=True, ax=ax)
ax.set_title('Distribuição das Notas de Satisfação')
ax.set_xlabel('Nota de Satisfação')
ax.set_ylabel('Frequência')

# 3️⃣ Salvar usando a função
salvar_grafico(fig,
               nome_arquivo='distribuicao_satisfacao',
               pasta_destino=r'D:\03 - Portfolios\portfolio_diego\analise_aneel_satisfacao_consumidor\imagens')
```

```python
def plot_histograma(df, coluna, nome_arquivo, pasta_destino):
    fig, ax = plt.subplots(figsize=(12, 6))
    sns.histplot(df[coluna], bins=20, kde=True, ax=ax)
    ax.set_title(f'Distribuição da coluna: {coluna}')
    ax.set_xlabel(coluna)
    ax.set_ylabel('Frequência')
    salvar_grafico(fig, nome_arquivo, pasta_destino)
```

```python
def plot_boxplot(df, coluna, nome_arquivo, pasta_destino):
    fig, ax = plt.subplots(figsize=(10, 6))
    sns.boxplot(x=df[coluna], ax=ax)
    ax.set_title(f'Boxplot da coluna: {coluna}')
    ax.set_xlabel(coluna)
    salvar_grafico(fig, nome_arquivo, pasta_destino)
```

```python
def plot_barplot(df, coluna_x, coluna_y, nome_arquivo, pasta_destino):
    fig, ax = plt.subplots(figsize=(12, 6))
    sns.barplot(x=df[coluna_x], y=df[coluna_y], ax=ax)
    ax.set_title(f'Gráfico de Barras: {coluna_x} vs {coluna_y}')
    ax.set_xlabel(coluna_x)
    ax.set_ylabel(coluna_y)
    salvar_grafico(fig, nome_arquivo, pasta_destino)
```

```python
def plot_scatterplot(df, coluna_x, coluna_y, nome_arquivo, pasta_destino):
    fig, ax = plt.subplots(figsize=(10, 6))
    sns.scatterplot(x=df[coluna_x], y=df[coluna_y], ax=ax)
    ax.set_title(f'Dispersão: {coluna_x} vs {coluna_y}')
    ax.set_xlabel(coluna_x)
    ax.set_ylabel(coluna_y)
    salvar_grafico(fig, nome_arquivo, pasta_destino)
```
